## 범주형 자료 탐색
* GENDER
* EDUCATION
* MARRIAGE
* PAY1 ~ PAY6
* DEFAULT

```python
df.GENDER.value_counts()
```

```python
"""
- 1 = 남성
- 2 = 여성
"""
df.GENDER.value_counts().sort_index()  
```

```python
"""
- 1 = 대학원 졸업
- 2 = 대학교 졸업
- 3 = 고등학교 졸업
- 4 = 기타

NOTE:
- 0, 5, 6: 데이터셋 소개에 언급되지 않은 카테고리
- 개수가 적으니 분석시 제외하는 것을 고려 할 수 있음
"""

df.EDUCATION.value_counts().sort_index()
```

```python
"""
- 1: 기혼
- 2: 미혼
- 3: 기타

NOTE:
- 0: 데이터셋 소개에 언급되지 않은 카테고리
- 마찬가지로 개수가 적으니 분석시 제외하는 것을 고려 할 수 있음
"""

df.MARRIAGE.value_counts().sort_index()
```

```python
"""
- -1 = 정상 지불
- 1 = 1개월 지연
- 2 = 2개월 지연
- ...
- 8 = 8개월 지연
- 9 = 9개월 이상 지연

NOTE:
- 데이터셋 소개 문서에 나온 것보다 1씩 작게 되어 있음
- PAY1 ~ PAY6 값들을 +1씩 올려줘야 함
"""

df.PAY1.value_counts().sort_index()
```

```python
"""
- 0 = 정상
- 1 = 연체
"""

df.DEFAULT.value_counts().sort_index()
```

```python
df.loc[
    (df['DEFAULT'] == 0) & (~df['PAY1'].isin([0, -1, -2])),
    ['DEFAULT', 'PAY1']
]

# 조건을 만족하는 행의 개수 > 정상이면서 PAY1은 연체된 데이터들이 왜 있는거지?
```

## 수치형 자료 탐색 (연속형)

```python
bill_amt_cols = [f"BILL_AMT{i}" for i in range(1, 7)]
pay_amt_cols = [f"PAY_AMT{i}" for i in range(1, 7)]

numeric_cols = ["LIMIT_BAL"] + bill_amt_cols + pay_amt_cols
```

```python
df[bill_amt_cols].head()
```

```python
df[pay_amt_cols].head()
```

```python
df[numeric_cols].head()
```

## 데이터의 중심

```python
# 평균
df[bill_amt_cols].mean()
```

```python
# 중앙값 (극단치에 robust)
df[bill_amt_cols].median()
```

```python
# 중앙값은 quantile로도 계산 가능
df[bill_amt_cols].quantile(0.5)
```

```python
# quantile은 동시에 여러 값 계산 가능
df[bill_amt_cols].quantile([0.0, 0.1, 0.5, 0.9, 1.0])
```

```python
# quantile(0.0): 최솟값
df[bill_amt_cols].min()
```

```python
# quantile(1.0): 최댓값
df[bill_amt_cols].max()
```

## 데이터의 분포

```python
# 범위
bill_amt_max = df[bill_amt_cols].max()
bill_amt_min = df[bill_amt_cols].min()
bill_amt_range = bill_amt_max - bill_amt_min
bill_amt_range
```

```python
# 표준편차
df[bill_amt_cols].std()
```

```python
# 사분위 범위
bill_amt_q3 = df[bill_amt_cols].quantile(0.75)
bill_amt_q1 = df[bill_amt_cols].quantile(0.25)
bill_amt_iqr = bill_amt_q3 - bill_amt_q1
bill_amt_iqr
```

## 상관계수

```python
# 지난 6개월 청구 금액의 상관계수
df[bill_amt_cols].corr().round(2)
```

```python
# 지난 6개월 선납 금액의 상관계수
df[pay_amt_cols].corr().round(2)
```

```python
# 수치형 자료 전체에 대한 상관계수
df[numeric_cols].corr().round(2)
```
